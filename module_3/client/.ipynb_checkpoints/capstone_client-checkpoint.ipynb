```python
promo_query = """
SELECT o.order_id, CONCAT(c.first_name, ' ', c.last_name) AS full_name, ot.pretip_total, c.customer_phone
FROM orders o
INNER JOIN order_totals ot ON o.order_id = ot.order_id
INNER JOIN customers c ON c.customer_id = o.customer_id
WHERE ot.pretip_total > 60;
 """

cursor.execute(promo_query)
promo_result = cursor.fetchall()
for i in promo_result:
    print(i)
```
